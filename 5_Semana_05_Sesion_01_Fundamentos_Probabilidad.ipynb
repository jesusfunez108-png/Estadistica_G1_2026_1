## **Fase 3: Fundamentos de Probabilidad en Python**

### 🌍 Contexto en Ingeniería
En el mundo de la ingeniería, ya sea civil, agrícola o agroindustrial, nos enfrentamos constantemente a situaciones inciertas. La **probabilidad** es el lenguaje matemático que nos permite cuantificar y modelar esta incertidumbre, ayudándonos a tomar decisiones informadas ante riesgos, variaciones en materiales o rendimientos, y la predicción de eventos futuros.

### 🔍 Conceptos Fundamentales
Para entender la probabilidad, es crucial dominar algunos conceptos básicos:

*   **Experimento Aleatorio:** Cualquier proceso cuyo resultado no se puede predecir con certeza, pero cuyos posibles resultados son conocidos (ej., el lanzamiento de un dado, la inspección de una viga).
*   **Espacio Muestral ($\Omega$):** El conjunto de todos los posibles resultados de un experimento aleatorio. Para la inspección de una viga, $\Omega = \{ \text{Aprobado, Fallo_Menor, Fallo_Crítico} \}$.
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Es decir, uno o más resultados específicos de interés (ej., el evento "la viga falla" podría ser $A = \{ \text{Fallo_Menor, Fallo_Crítico} \}$).

### 📊 Ecuaciones de Probabilidad
Existen dos enfoques principales para calcular probabilidades:

*   **Probabilidad Teórica (Regla de Laplace):** Cuando todos los resultados de un espacio muestral finito son igualmente probables, la probabilidad de un evento $A$ se calcula como:
    
    $$ P(A) = \frac{\text{Número de casos favorables a } A}{\text{Número total de casos posibles}} $$

*   **Probabilidad Empírica (Frecuencia Relativa):** Basada en la observación de experimentos o datos históricos. Si un experimento se repite $n$ veces y el evento $A$ ocurre $k$ veces, la probabilidad empírica de $A$ es:
    
    $$ P(A) \approx \frac{\text{Número de veces que ocurre } A}{\text{Número total de ensayos}} = \frac{k}{n} $$

---
Ahora, aplicaremos estos conceptos teóricos a la práctica, simulando datos en Python para calcular probabilidades empíricas y observar cómo se alinean con nuestra comprensión.

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## **Fase 4: Probabilidad Condicional e Independencia en Python**

### 💡 Introducción Práctica
Tras calcular las probabilidades empíricas generales, surge una pregunta fundamental en ingeniería: ¿Cómo cambia la probabilidad de un evento si ya sabemos que otro evento ha ocurrido? Por ejemplo, ¿la probabilidad de que una viga tenga un `Fallo_Critico` es la misma independientemente del proveedor que la suministró? Investigar estas relaciones nos permite tomar decisiones más informadas, como seleccionar proveedores de mayor calidad o ajustar procesos de control.

### 🔑 Conceptos Clave

*   **Probabilidad Condicional:** Es la probabilidad de que un evento $A$ ocurra, *dado que* otro evento $B$ ya ha ocurrido. Nos permite actualizar nuestra estimación de incertidumbre basada en nueva información. Se denota como $P(A|B)$.
*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En términos de control de calidad, significa que la calidad del producto no está influenciada por factores como el proveedor, la línea de producción, etc. Si no son independientes, decimos que son **dependientes**.

### 📝 Ecuaciones en LaTeX

*   **Probabilidad Condicional:**
    $$ P(A|B) = \frac{P(A \cap B)}{P(B)} \quad \text{donde } P(B) > 0 $$
    Esta fórmula nos dice que la probabilidad de $A$ dado $B$ es la probabilidad de que ambos $A$ y $B$ ocurran, dividida por la probabilidad de $B$.

*   **Condición de Independencia:**
    Dos eventos $A$ y $B$ son independientes si:
    $$ P(A|B) = P(A) $$
    O, equivalentemente (y más comúnmente utilizada para verificar independencia):
    $$ P(A \cap B) = P(A) \cdot P(B) $$
    Si estas condiciones no se cumplen, los eventos son dependientes.

---
A continuación, utilizaremos `pandas` en Python para construir tablas de contingencia cruzada, lo que nos facilitará el cálculo de estas probabilidades condicionales y nos ayudará a evaluar la independencia entre el estado de inspección y el proveedor.

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
